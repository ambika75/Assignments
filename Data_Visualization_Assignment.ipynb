```python
#importing libraries
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn import datasets
%matplotlib inline
```

```python
#Loading Dataset
iris = datasets.load_iris()
X, Y = iris['data'], iris['target']
X.shape, Y.shape
X
```

```python
iris.target_names
```

```python
#Decomposing dimension of data
pca_X=decomposition.PCA(n_components=3)
X= pca_X.fit_transform(X)
X
```

```python
# Plotting the data
x = np.array(X[:,0])
y = np.array(X[:,1])
z = np.array(X[:,2])
fig = plt.figure(1)
ax = Axes3D(fig,proj_type= "ortho")
ax.scatter(x, y, z,c=Y,cmap=plt.cm.autumn,depthshade=False)
```
